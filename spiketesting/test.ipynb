```python
import numpy as np

def sliding_window(data, window_size, future_steps):
    """
    Sliding window 함수.
    
    Parameters:
        - data: 시계열 데이터 (리스트 또는 배열)
        - window_size: 한 창의 크기
        - future_steps: 예측할 미래의 시간 간격 (window 단위)
        
    Returns:
        - forecasts: 주어진 데이터에 대한 미래 예측
    """
    forecasts = []
    
    # 시작 인덱스 설정
    start_index = 0
    
    # 마지막 인덱스 계산
    end_index = len(data) - window_size + 1
    
    # future_steps 만큼 반복하여 예측 수행
    for _ in range(future_steps):
        # 현재 창에서 종가 가져오기
        window_data = data[start_index:start_index+window_size]
        close_prices = [row[-1] for row in window_data]
        
        # 다음 창 예측
        next_forecast = np.mean(close_prices)
        forecasts.append(next_forecast)
        
        # 다음 창으로 이동
        start_index += 1
    
    return forecasts

# 예시 데이터
time_series_data = [
    [1, 2, 3, 4],  # 시간 1
    [2, 3, 4, 5],  # 시간 2
    [3, 4, 5, 6],  # 시간 3
    # 이하 계속...
]

# sliding window 함수 호출
window_size = 336  # 336시간
future_steps = 63  # 63일
forecasts = sliding_window(time_series_data, window_size, future_steps)

print("63일 동안의 종가 예측:")
print(forecasts)
```
